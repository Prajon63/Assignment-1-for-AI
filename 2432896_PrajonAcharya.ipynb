```python
#Task - 3 - Outlier Detection:
#Tasks:
#1. Identify outlier countries in South Asia based on their Score and GDP per Capita.

# Calculate the IQR for Score and Log GDP per capita
Q1_score = south_asian_df['score'].quantile(0.25)
Q3_score = south_asian_df['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

Q1_gdp = south_asian_df['Log GDP per capita'].quantile(0.25)
Q3_gdp = south_asian_df['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

# Define outlier thresholds for Score and Log GDP per capita using IQR
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# Detect outliers for Score and Log GDP per capita
outliers_score = south_asian_df[(south_asian_df['score'] < lower_bound_score) | (south_asian_df['score'] > upper_bound_score)]
outliers_gdp = south_asian_df[(south_asian_df['Log GDP per capita'] < lower_bound_gdp) | (south_asian_df['Log GDP per capita'] > upper_bound_gdp)]

# Display outliers
print("Outlier countries based on Score:")
print(outliers_score[['Country name', 'score']])

print("\nOutlier countries based on Log GDP per capita:")
print(outliers_gdp[['Country name', 'Log GDP per capita']])

```

```python
#2. Define outliers using the 1.5  IQR rule.
  # Calculate Q1, Q3, and IQR for 'Score' and 'Log GDP per capita'
Q1_score = south_asian_df['score'].quantile(0.25)
Q3_score = south_asian_df['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

Q1_gdp = south_asian_df['Log GDP per capita'].quantile(0.25)
Q3_gdp = south_asian_df['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

# Define lower and upper bounds using the 1.5 * IQR rule for both Score and Log GDP per capita
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# Detect outliers for Score and Log GDP per capita using the 1.5 * IQR rule
outliers_score = south_asian_df[(south_asian_df['score'] < lower_bound_score) |
(south_asian_df['score'] > upper_bound_score)]

outliers_gdp = south_asian_df[(south_asian_df['Log GDP per capita'] < lower_bound_gdp) |
(south_asian_df['Log GDP per capita'] > upper_bound_gdp)]

# Display outliers
print("Outlier countries based on Score using 1.5 * IQR rule:")
print(outliers_score[['Country name', 'score']])

print("\nOutlier countries based on Log GDP per capita using 1.5 * IQR rule:")
print(outliers_gdp[['Country name', 'Log GDP per capita']])
```
